```python
# Load the Dataset, we get only the data without the tolerances [0:3]
NO2_road_df = pd.read_csv("NO2_tables_roadside.csv").iloc[:,:3]
NO2_urban_df = pd.read_csv("NO2_tables_urban.csv").iloc[:,:3]

PM10_road_df = pd.read_csv("PM10_tables_roadside.csv").iloc[:,:3]
PM10_urban_df = pd.read_csv("PM10_tables_urban.csv").iloc[:,:3]

PM25_road_df = pd.read_csv("PM2.5_tables_roadside.csv").iloc[:,:3]
PM25_urban_df = pd.read_csv("PM2.5_tables_urban.csv").iloc[:,:3]

O3_road_df = pd.read_csv("O3_tables_roadside.csv").iloc[:,:3]
O3_urban_df = pd.read_csv("O3_tables_urban.csv").iloc[:,:3]
```

```python
urban_list = [NO2_urban_df, PM10_urban_df, PM25_urban_df, O3_urban_df]

road_list = [NO2_road_df, PM10_road_df, PM25_road_df, O3_road_df]

urban_dfs = [df.set_index(['Year', 'Site']) for df in urban_list]
road_dfs = [df.set_index(['Year', 'Site']) for df in road_list]
```

```python
pd.concat(urban_dfs, axis=1).reset_index().dropna()
```

```python
pd.concat(road_dfs, axis=1).reset_index().dropna()
```

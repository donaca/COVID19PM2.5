##### Loading of COVID dataset for UK

```python
UK_covid_df = pd.read_csv("lahbtablesweek20finalcodes.csv")
UK_covid_df = UK_covid_df[['Area name', 'Cause of death', 'Number of deaths']]

# What cause of death shall we consider? All or only Covid? TODO
UK_covid_df = UK_covid_df[UK_covid_df['Cause of death'] == 'COVID 19']

# Cumulative sum of death by area since we discard the 'place' of death previously
UK_covid_df = UK_covid_df.groupby('Area name', as_index=False).agg({'Number of deaths': 'sum'})
UK_covid_df.columns = ['site', '# death']

UK_covid_df.to_csv('UK_covid.csv', index=False)
```
